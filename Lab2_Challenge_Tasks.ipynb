```python
# TASK 1 

surname = "DEVERA"
seed_digit = "0"

result = []

for letter in surname:
    if letter in ["A", "E", "I", "O", "U"]:
        result.append(seed_digit)
    else:
        result.append(letter)

final_output = "".join(result)

print("Original Surname:", surname)
print("Redacted Output:", final_output)
```

```python
# TASK 2 

surname = "DEVERA"
seed_value = 0

name_length = len(surname)
first_char = surname[0]
ascii_value = ord(first_char)

hash_result = ascii_value * seed_value - name_length

print("First Letter ASCII:", ascii_value)
print("Verification Hash:", hash_result)
```

```python
# TASK 3 

stored_password = "REDTS143"
attempt_count = 0
access_granted = False

while not access_granted:
    entered_password = input("Enter password: ")
    attempt_count += 1

    if entered_password == stored_password:
        print("Access Granted!")
        access_granted = True
    else:
        print("Incorrect. System Locked.")

print("Attempts:", attempt_count)
```
